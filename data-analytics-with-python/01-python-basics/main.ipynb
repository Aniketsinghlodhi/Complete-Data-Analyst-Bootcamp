```python
## case sensitivity - Python is case sensitive language. So 'a' and 'A' are different variables.
name = "Aniket"
Name = "Lodhi"  
print(name)
print(Name)
```

## Indentation in python is used to define the structure and hierarchy of the code .unlike many other programming languages that use curly {} braces or keywords to denote code blocks, Python relies on indentation levels to indicate which statements belong to which blocks. this mean that the amount of whitespace (spaces or tabs) at the beginning of a line is significant and determines the grouping of statements.


```python
## indentation 
## python used indentation to define the block of code. consistent use of spaces (commonly 4 spaces) or tabs is essential to maintain the structure and readability of the code.
age = 18
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")  
## in the above code indentation is used to define the block of code for if and else statements.
```

```python
## Line condinuation
## In Python, line continuation is used to split long lines of code into multiple lines for better readability. This can be achieved using the backslash (\) character at the end of a line, indicating that the statement continues on the next line.
total = 1 + 2 + 3 + \
        4 + 5 + 6 + \
        7 + 8 + 9
print("Total:", total)  
```

```python
## Multiple statements on a single line
## In Python, you can write multiple statements on a single line by separating them with a semicolon (;). However, it's generally recommended to use this feature sparingly, as it can reduce code readability.
x = 5; y = 10; z = x + y; print("Sum:", z)
```
